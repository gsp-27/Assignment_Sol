```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# unvectorised version
n = 10
ones = 6
zeroes = 4
thetas = np.arange(0,1.01,0.01)
# likelihood for each of the thetas is (theta ** ones) * (1-theta ** zeroes)
likelihood = list()
for theta in thetas:
    likelihood.append((theta ** ones) * ((1-theta) ** zeroes))
likelihood = np.array(likelihood)
index = likelihood == np.max(likelihood)
max_theta = thetas[index]
print max_theta
plt.scatter(thetas, likelihood, s=20)
```

```python
# vectorised version
def MLE(ones, zeroes, thetas):
    likelihood = (thetas ** ones) * ((1-thetas)**zeroes)
    index = likelihood == np.max(likelihood)
    max_theta = thetas[index]
    return max_theta, likelihood
```

```python
theta5, like5 = MLE(3,2,thetas)
print theta5
plt.plot(like5)
```

```python
theta100, like100 = MLE(60,40,thetas)
print theta100
plt.plot(like100)
```

```python
theta10, like10 = MLE(5,5,thetas)
print theta10
plt.plot(like10)
```

```python
# Maximum aposteriori estimation
beta33 = 0.0333
ones = 6
zeros = 4
n = 10
thetas = np.arange(0, 1.01, 0.01)
posterior = list()
for theta in thetas:
    prior = ((theta ** 2) * ((1 - theta)**2)) / beta33
    likelihood = (theta ** ones) * ((1 - theta)**zeros)
    posterior.append(likelihood *  prior)
posterior = np.array(posterior)
plt.plot(posterior)
maxtheta = posterior == np.max(posterior)
print thetas[maxtheta]
```

```python
# vectorised version it basically involves elementwise multiplication
prior = ((thetas ** 2) * ((1 - thetas) ** 2)) / beta33
likelihood = (thetas ** ones) * ((1 - thetas) ** zeros)
posterior = likelihood * prior
plt.plot(posterior)
max_theta_index = posterior == np.max(posterior)
max_theta = thetas[max_theta_index]
print max_theta
```
